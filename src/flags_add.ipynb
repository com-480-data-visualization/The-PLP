```python
import pandas as pd

# Load the CSV files into dataframes
countries_df = pd.read_csv('../assets/data/basic_dataset/countries_flag.csv')
guns_death_df = pd.read_csv('../assets/data/computed_dataset/partial_computed_csv.csv')


# Merge the dataframes based on country names
merged_df = pd.merge(countries_df, guns_death_df, left_on='country', right_on='country')
#merged_df = merged_df.drop('Country', axis=1)
display(merged_df.head())
```

```python
countries_capital_df = pd.read_csv('../assets/data/basic_dataset/capital.csv', encoding='cp1252', sep=';')
```

```python
merged_df_2 = pd.merge(merged_df, countries_capital_df, left_on='alpha3', right_on='CODE')
display(merged_df_2.head())
```

```python
columns_to_drop = ['images_file', 'alpha2', 'iso_31662','intermediateregion','NOM_ALPHA','NOM_LONG','NOM','CODE','intermediateregioncode','subregioncode','regioncode','subregion','ARTICLE']  # Remplacez par les noms des colonnes que vous souhaitez supprimer
merged_df_2 = merged_df_2.drop(columns=columns_to_drop)


display(merged_df_2.head())
```

```python

merged_df_2.to_csv('../assets/data/computed_dataset/full_dataset.csv', index=False)
```
